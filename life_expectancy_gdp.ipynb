### 1. Importing all the libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

### 2. Importing and diagnosing the data

```python
# import the data and print first few rows
df = pd.read_csv("all_data.csv")
df.head()
```

```python
# renaming the "too long" column names
df = df.rename(columns={'Life expectancy at birth (years)': 'Life_expectancy'})
```

```python
df.describe()
```

```python
df.info()
```

```python
for col in df:
    print(df[col].unique())
```

### 3. Creating sub-dataframes for each country separatley

```python
# creating sub-dataframes for each country
chile = df[(df.Country == "Chile")]
china = df[(df.Country == "China")]
germany = df[(df.Country == "Germany")]
mexico = df[(df.Country == "Mexico")]
usa = df[(df.Country == "United States of America")]
zimbabwe = df[(df.Country == "Zimbabwe")]
```

### 4. Plotting some graphs showing GDP and Life expectancies for the countries in our dataset 

```python
# Life expectancy in countries through years
sns.scatterplot(x = "Year", y = "Life_expectancy", hue = "Country", palette = "bright", data = df)
plt.title("Life expectancy in countries from 2000 - 2015")
plt.show()
plt.clf()
```

```python
# GDP in countries through years
sns.scatterplot(x = "Year", y = "GDP", hue = "Country", palette = "bright", data = df)
plt.title("GDP in countries from 2000 - 2015")
plt.show()
plt.clf()
```

```python
plt.rcParams['figure.figsize'] =(10, 7)
plt.suptitle('Countrys Life Expectancy through Years')

# Subplot 1 - Chile Life Expectancy
plt.subplot(2, 3, 1)
plt.plot(chile.Year, chile.Life_expectancy, color="blue")
plt.title("Chile")

# Subplot 2 - Chinas Life Expectancy
plt.subplot(2, 3, 2)
plt.plot(china.Year, china.Life_expectancy, color="orange")
plt.title("China")

# Subplot 3 - Germanys Life Expectancy
plt.subplot(2, 3, 3)
plt.plot(germany.Year, germany.Life_expectancy, color="green")
plt.title("Germany")

# Subplot 4 - Mexicos Life Expectancy
plt.subplot(2, 3, 4)
plt.plot(mexico.Year, mexico.Life_expectancy, color="red")
plt.title("Mexico")

# Subplot 5 - USA Life Expectancy
plt.subplot(2, 3, 5)
plt.plot(usa.Year, usa.Life_expectancy, color="purple")
plt.title("USA", )

# Subplot 6 - Zimbabwes Life Expectancy
plt.subplot(2, 3, 6)
plt.plot(zimbabwe.Year, zimbabwe.Life_expectancy, color="brown")
plt.title("Zimbabwe")

plt.suptitle("Life Expectancy")
plt.tight_layout()

plt.show()
plt.clf()
```

```python
plt.rcParams['figure.figsize'] =(10, 7)
plt.suptitle('Countrys Life Expectancy against GDP')

plt.subplot(2, 3, 1)
plt.scatter(chile.Life_expectancy, chile.GDP, color="blue")
plt.title("Chile")

plt.subplot(2, 3, 2)
plt.scatter(china.Life_expectancy, china.GDP, color="orange")
plt.title("China")

plt.subplot(2, 3, 3)
plt.scatter(germany.Life_expectancy, germany.GDP, color="green")
plt.title("Germany")

plt.subplot(2, 3, 4)
plt.scatter(mexico.Life_expectancy, mexico.GDP, color="red")
plt.title("Mexico")

plt.subplot(2, 3, 5)
plt.scatter(usa.Life_expectancy, usa.GDP, color="purple")
plt.title("USA")

plt.subplot(2, 3, 6)
plt.scatter(zimbabwe.Life_expectancy, zimbabwe.GDP, color="brown")
plt.title("Zimbabwe")

plt.tight_layout()
plt.show()
plt.clf()
```

### Conclusion

- We can see that the developing and developed countries have similar Life expectancy from 70-80 years, slightly going up.
- The outlier in  Life expectancy is Zimbabwe with Life expectancy lower than 60 years, but going up from 45 to 60 in years from 2005 - 2015.

- The high GDP is not the reason for the high Life expectancy.

- The outlier in GDP in USA, with more than 1.5 trillions $.
- Chinas GDP is going up in last few years.
